```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

results_path = "./results/5_final_results/"

sns.set_theme()
```

```python
results = pd.read_csv(results_path + '/mean.csv')
```

```python
p = sns.histplot(data=results, x="botwiki-verified", binwidth=0.01).set(title='Tested on botwiki-verified', xlabel='AUC scores', ylabel='Frequency', xlim=(0.0,1.0))
plt.show()
p = sns.histplot(data=results, x="midterm-18", binwidth=0.01).set(title='Tested on midterm-18', xlabel='AUC scores', ylabel='Frequency', xlim=(0.0,1.0))
plt.show()
p = sns.histplot(data=results, x="gilani-17", binwidth=0.01).set(title='Tested on gilani-17', xlabel='AUC scores', ylabel='Frequency', xlim=(0.0,1.0))
plt.show()
p = sns.histplot(data=results, x="cresci-rtbust", binwidth=0.01).set(title='Tested on cresci-rtbust', xlabel='AUC scores', ylabel='Frequency', xlim=(0.0,1.0))
plt.show()
p = sns.histplot(data=results, x="5-fold_cross-validation", binwidth=0.01).set(title='5-fold cross-validation', xlabel='AUC scores', ylabel='Frequency', xlim=(0.0,1.0))
plt.show()
p = sns.histplot(data=results, x="spearman_r", binwidth=0.01).set(title='Correlation with Botometer', xlabel='Spearman\'r', ylabel='Frequency', xlim=(-1.0,1.0))
plt.show()
```
