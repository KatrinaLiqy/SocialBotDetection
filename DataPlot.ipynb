```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

results_path = "./results/5_final_results/"

sns.set_theme()
```

```python
results = pd.read_csv(results_path + '/mean.csv')
```

```python
graph_settings = {"xlabel": "AUC", "ylabel": "Frequency", "xlim": (-0,1), "xticks": [0.25, 0.5, 0.75, 1.0]}

p = sns.histplot(data=results, x="botwiki-verified", binwidth=0.02).set(title='Tested on botwiki-verified', **graph_settings)
plt.show()
p = sns.histplot(data=results, x="midterm-18", binwidth=0.02).set(title='Tested on midterm-18', **graph_settings)
plt.show()
p = sns.histplot(data=results, x="gilani-17", binwidth=0.02).set(title='Tested on gilani-17', **graph_settings)
plt.show()
p = sns.histplot(data=results, x="cresci-rtbust", binwidth=0.02).set(title='Tested on cresci-rtbust', **graph_settings)
plt.show()
p = sns.histplot(data=results, x="5-fold_cross-validation", binwidth=0.02).set(title='5-fold cross-validation', **graph_settings)
plt.show()
p = sns.histplot(data=results, x="spearman_r", binwidth=0.04).set(title='Correlation with Botometer', xlabel='Spearman\'s r', ylabel='Frequency', xlim=(-1.2,1.2))
plt.show()
```
